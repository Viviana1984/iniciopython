Crear una variable para almacenar el nombre del usuario y otra para su edad, para mostar en la pantalla un mensaje de bienvenida

Crear una función que determine si el usuario es mayor de edad

```python
def es_mayor_de_edad(edad):
  if edad >=18:
    return "Eres mayor de edad"
  else:
    return "Eres menor de edad"
print(es_mayor_de_edad(edad))
```

Los bucles en python se utilizan para ejecutar un bloque de código de forma repetida. El bucle for, se utiliza cuando sabemos donde va a terminar la repetición. El bucle while, se utiliza cuando NO sabemos donde va terminar

```python
for recorrer in range(5):
  print(recorrer)
```
